```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.special import factorial
```

Otra forma de calcular el valor de π es:

$$\pi= 3*arcos(\frac{1}{2})$$

La sumatoria de arcos($\frac{1}{2}$):

$$\pi=3* (\frac{\pi}{2} - \sum_{n=0}^{\infty} (\frac{(2n)!}{(2^{n}n!)^{2}}\cdot \frac{(1/2)^{2n+1}}{2n+1})$$

moviendo π para la derecha y la sumatoria a la izquierda para obtener ambos positivos:
$$ \pi= 6* \sum_{n=0}^{\infty} (\frac{(2n)!}{(2^{n}n!)^{2}}\cdot \frac{(1/2)^{2n+1}}{2n+1}) $$

Expandiendo la serie se obtiene:

$$ \pi= 3 + \frac{1}{8} + \frac{9}{640} + ...etc  $$


```python
#tomamos el primer valor de la suma
Taylor_pi=[3]

#Realizamos la sumatoria hasta llegar al término n=10
for n in range(1,10):
    Taylor_pi.append((Taylor_pi[n-1]+6*(math.factorial(2*n)*(1/2)**(2*n+1)/((2**(n)*math.factorial(n))**2*(2*n+1)))))
#la convergencia con el valor original de pi es conveniente graficar de forma logaritmica
plt.loglog(np.arange(10), Taylor_pi, label='Aproximacion $\pi$')
plt.loglog(np.arange(10), np.full(10, np.pi), label = '$\pi$ real')

#Para observar mejor el grafico es posible realizar un acercamiento al grafico
plt.ylim(np.pi - 0.005*np.pi, np.pi + 0.005*np.pi)
plt.legend()
plt.show()
```

podemos ver que a partir del cuarto término ya converge lo más cerca al valor original de π. la aproximación es certera ya que solo ocupamos 10 terminos de la sumatoria.

###comparación de ambas aproximaciones(vista en clases).

```python
#Aproximacion de pi vista en clases pero recortado a 10 para igualar con nuestra aproximación.
pi_1 = [4]
pi_2 = [3]

for n in range(1, 10):
    pi_1.append(pi_1[n-1] + (-1)**n*4/(2*n + 1))
    pi_2.append(pi_2[n-1] + (-1)**(n + 1)*4/(2*n*(2*n + 1)*2*(n+1)))

#Ahora se obtiene las graficas de las 3 aproximaciones y el valor real de pi
plt.loglog(np.arange(10), Taylor_pi, label='Aproximacion $\pi$')
plt.loglog(np.arange(10), pi_1, label='pi_1')
plt.loglog(np.arange(10), pi_2, label='pi_2')
plt.loglog(np.arange(10), np.full(10, np.pi), label = '$\pi$ real')
plt.ylim(np.pi - 0.005*np.pi, np.pi + 0.005*np.pi)
plt.legend()
plt.show()
```

la aproximación que obtuvimos arriba con la serie de taylor es la que converge más rápido al valor real de π, ya que ocupamos pocos términos de la sumatoria y a partir del cuarto término ya es muy cercano al valor original.
