En la clase pasada introdujimos dos funciones que calculan el valor aproximado de $\pi$.

* Escriba dos funciones que calculen $\pi$ usando cada una de esas aproximaciones respectivamente. Cada función debe tomar el error relativo como input y regresar el valor aproximado de $\pi$ como output.

* Luego calcule este valor de $\pi$ para diez errores relativos entre $0.1$ y $0.01$. Mida el tiempo que tarda cada uno para cada función y grafique el tiempo de ambas en función del error relativo.

```python
import time
import matplotlib.pyplot as plt

# Función para calcular pi utilizando la fórmula de Leibniz
def calcular_pi_leibniz(error_relativo):
    pi_aproximado = 0
    n = 0
    denominador = 1
    while True:
        termino = 4 / denominador
        if n % 2 == 0:
            pi_aproximado += termino
        else:
            pi_aproximado -= termino
        n += 1
        denominador += 2
        # Calcular el error relativo en cada iteración
        error_actual = abs(pi_aproximado - 3.14159265358979323846) / 3.14159265358979323846
        if error_actual < error_relativo:
            break
    return pi_aproximado

# Función para calcular pi utilizando la fórmula más rápida
def calcular_pi_mas_rapido(error_relativo):
    pi_aproximado = 3
    n = 1
    denominador = 2
    while True:
        termino = 4 / (denominador * (denominador + 1) * (denominador + 2))
        if n % 2 == 1:
            pi_aproximado += termino
        else:
            pi_aproximado -= termino
        n += 1
        denominador += 2
        # Calcular el error relativo en cada iteración
        error_actual = abs(pi_aproximado - 3.14159265358979323846) / 3.14159265358979323846
        if error_actual < error_relativo:
            break
    return pi_aproximado

# Lista de errores relativos para los cuales calcularemos pi
errores_relativos = [0.1, 0.09, 0.08, 0.07, 0.06, 0.05, 0.04, 0.03, 0.02, 0.01]

# Listas para almacenar los tiempos de ejecución de cada función
tiempos_leibniz = []
tiempos_mas_rapido = []

# Calcular pi y medir el tiempo para cada error relativo usando ambas fórmulas
for error_relativo in errores_relativos:
    inicio_tiempo = time.time()
    calcular_pi_leibniz(error_relativo)
    fin_tiempo = time.time()
    tiempos_leibniz.append(fin_tiempo - inicio_tiempo)

    inicio_tiempo = time.time()
    calcular_pi_mas_rapido(error_relativo)
    fin_tiempo = time.time()
    tiempos_mas_rapido.append(fin_tiempo - inicio_tiempo)

# Graficar los tiempos de ejecución en función del error relativo
plt.plot(errores_relativos, tiempos_leibniz, label='Fórmula de Leibniz')
plt.plot(errores_relativos, tiempos_mas_rapido, label='Fórmula más rápida')
plt.xlabel('Error Relativo')
plt.ylabel('Tiempo de Ejecución (segundos)')
plt.legend()
plt.title('Tiempo de Ejecución vs. Error Relativo para el Cálculo de Pi')
plt.show()
```
