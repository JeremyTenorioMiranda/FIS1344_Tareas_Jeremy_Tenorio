Considere la función $f(x) = 4 x^3 - 3x^2 + x -5$. Use sólo dos puntos para interpolar usando una spline cúbica entre $x =-1$ y $x = 1$. Haga lo mismo con 4 puntos equidistantes. Explique por qué la spline cúbica es diferente del polinomio original si ambos son polinomios cúbicos.

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

# Define la función original
def f(x):
    return 4 * x**3 - 3 * x**2 + x - 5

# Puntos de interpolación
x_points_2 = [-1, 1]  # Dos puntos
y_points_2 = [f(x) for x in x_points_2]

x_points_4 = np.linspace(-1, 1, 4)  # Cuatro puntos equidistantes
y_points_4 = [f(x) for x in x_points_4]

# Crear splines cúbicos
spline_2 = CubicSpline(x_points_2, y_points_2)
spline_4 = CubicSpline(x_points_4, y_points_4)

# Valores de x para graficar
x_plot = np.linspace(-1, 1, 100)

# Calcular los valores de la spline cúbica en esos puntos
y_spline_2 = spline_2(x_plot)
y_spline_4 = spline_4(x_plot)

# Gráficos
plt.figure(figsize=(10, 6))
plt.plot(x_plot, f(x_plot), label='Función original')
plt.plot(x_points_2, y_points_2, 'ro', label='Dos puntos de interpolación')
plt.plot(x_points_4, y_points_4, 'go', label='Cuatro puntos de interpolación')
plt.plot(x_plot, y_spline_2, 'b--', label='Spline cúbico (2 puntos)')
plt.plot(x_plot, y_spline_4, 'g--', label='Spline cúbico (4 puntos)')
plt.legend()
plt.title('Interpolación de una función con splines cúbicos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

La razón por la que la spline cúbica es diferente del polinomio original, es la forma en que se construyen y ajustan. La spline cúbica se adapta mejor a la curva porque utiliza varios polinomios cúbicos pequeños en vez de uno solo. Los polinomios cúbicos únicos pueden tener oscilaciones no deseadas y no proporcionar una interpolación suave, especialmente cuando se trata de datos ruidosos. La spline cúbica, al dividir el intervalo en segmentos más pequeños y ajustar polinomios separados a cada uno, puede proporcionar una interpolación más precisa y suave que sigue mejor la forma de la curva original.
